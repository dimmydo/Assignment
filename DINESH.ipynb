```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_excel('C:/Users/Zion_2232/Desktop/fuel-distance.xlsx')
```

```python
df
```

```python
df=pd.DataFrame(df,columns=['FUEL','DISTANCE'])
```

```python
df.plot.scatter(x='FUEL',y='DISTANCE')
```

```python
from sklearn.model_selection import train_test_split
X=df[['FUEL']]
Y=df[['DISTANCE']]
```

```python
X
```

```python
Y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
lm.fit(xtrain,ytrain)
```

```python
predictions=lm.predict(xtest)
predictions
```

```python
predictions=pd.DataFrame(predictions,columns=['predictions'])
predictions
```
